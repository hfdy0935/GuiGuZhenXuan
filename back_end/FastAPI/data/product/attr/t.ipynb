```python
import random
import json
data = ''
with open('./attrCategory.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

result = data

for index1, c1 in enumerate(data):

    for index2, t in enumerate(c1['contents']):

        c2 = t[0]

        c3 = t[1:]

        for index3, tt in enumerate(c3, start=1):
            tmp = []

            names = random.sample(list(attributes.keys()), random.randint(5, 25))

            for name in names:
                s = [i + str(random.randint(0,4)) for i in attributes[name]]
                values = random.sample(s, random.randint(3, 8))
                tmp.append({name: values})
            result[index1]['contents'][index2][index3] = {
                tt: tmp
            }
```

```python
with open('./attrList.json','w', encoding = 'utf-8') as f:
    json.dump(result, f, ensure_ascii=False,indent=2)
```
