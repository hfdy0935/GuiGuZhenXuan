```python
# 获取随机属性列表
def get_props_list():


    props_list = []

    num_dicts = random.randint(3,8)


    properties = random.sample(["颜色", "材质", "价格", "产地", "销量", "尺码", "款式", "品牌", "适用性别", "适用季节",


                                "包装方式", "年龄段", "功能", "适用场景", "工艺", "重量"], num_dicts)


    tags = ["时尚", "经典", "舒适", "高质量", "多功能",


            "清新", "个性化", "环保", "简约", "奢华", "红色", "黑色", "蓝色", '白色', '绿色', '橙色', '紫色', '黄色', '灰色', '纯棉', '优质', '高性价比', '天然', '养生', '节约', '保暖', '气质']


    for i in range(1, num_dicts + 1):


        prop_dict = {}


        prop_dict["序号"] = i


        prop_dict["属性"] = properties[i-1]


        # 生成随机标签


        num_tags = random.randint(4, 13)  # 随机生成标签数量，范围为6到10


        tag = random.sample(tags, num_tags)


        prop_dict["标签"] = tag


        # 生成随机标签颜色


        colors = [random.randint(0, 4)


                  for _ in range(num_tags)]  # 随机生成标签颜色，范围为0到4


        prop_dict["标签颜色"] = colors


        props_list.append(prop_dict)


    return props_list
```

```python
# 汇总之前三个部分，生成一个三级分类对应的spu
def get_spu_info():
    info_list = get_spu_name_description_list()
    for index,info in enumerate(info_list):
        info_list[index]['images'] = get_images_list()
        info_list[index]['props'] = get_props_list()
    return info_list
```

```python
# 初始化三种
# spuListData
# skuInSpuBeforeAdd
# addSkuList

def function():
    # 每个三级分类
    f = open('./attrCategory.json', 'r', encoding='utf-8')
    original_data = json.load(f)
    f.close()
    f = open('./attrCategory.json', 'r', encoding='utf-8')
    original_data1 = json.load(f)
    f.close()
    f = open('./attrCategory.json', 'r', encoding='utf-8')
    original_data2 = json.load(f)
    f.close()

    if not os.path.exists('./spuListData'):
        os.mkdir('./spuListData')
    if not os.path.exists('./skuInSpuBeforeAdd'):
        os.mkdir('./skuInSpuBeforeAdd')
    if not os.path.exists('./addSkuList'):
        os.mkdir('./addSkuList')
        
    for index1, c1 in enumerate(original_data):
        c1_title = c1['title'].replace('/','_')
        c1_sku = original_data1[index1]
        c1_add_sku = original_data2[index1]
        json_path = os.path.join('./spuListData', f'{index1+1}_{c1_title}.json')
        json_path1 = os.path.join('./skuInSpuBeforeAdd', f'{index1+1}_{c1_title}.json')
        json_path2 = os.path.join('./addSkuList', f'{index1+1}_{c1_title}.json')
        for index2, c23 in enumerate(c1['contents']):
            for index3, c3 in enumerate(c23[1:], start=1):
                t1 = get_spu_info() # 三级分类下的所有spu
                c1['contents'][index2][index3] = {
                    c3:  t1
                }
                # with open(json_path, 'w', encoding='utf-8') as f:
                #     json.dump(c1, f,ensure_ascii = False,indent = 2)

                    
                    
                # s2是三级分类下的一个商品，还要遍历生成三级分类下对应数量的所有商品  
                s2 = []
                # 遍历每个spu，对每个spu生成随机sku选项
                for i in t1: 
                    s2t: dict = get_sku_info()
                    
                    t = i['images']
                    s2t['images'] = t
                    tt = [0]*len(t)
                    tt[random.randint(0, len(t) - 1)] = 1
                    s2t['isDefault'] = tt
                    
                    s2.append(s2t)
                        
                c1_sku['contents'][index2][index3] = {
                    c3: s2
                }
                # with open(json_path1, 'w', encoding='utf-8') as f:
                #     json.dump(c1_sku, f,ensure_ascii = False,indent = 2)
                    
                    
                s3 = []
                for index,item in enumerate(t1):
                    s3.append({
                        '序号': index+1,
                        'spuName': item['spuName'],
                        'skuInfo': [],
                        # 'isShow': False
                    })
                c1_add_sku['contents'][index2][index3] = {
                    c3: s3
                }
                # with open(json_path2, 'w', encoding='utf-8') as f:
                #     json.dump(c1_add_sku, f, ensure_ascii=False, indent=2)
        with open(json_path, 'w', encoding='utf-8') as f:
            json.dump(c1, f,ensure_ascii = False,indent = 2)
        with open(json_path1, 'w', encoding='utf-8') as f:
            json.dump(c1_sku, f,ensure_ascii = False,indent = 2)
        with open(json_path2, 'w', encoding='utf-8') as f:
            json.dump(c1_add_sku, f, ensure_ascii=False, indent=2)
        
function()
```
