```python
# 初始化所有角色
import json
import time
import random
import datetime
import os
```

```python
# 生成随机时间
def generate_create_time():
    year = random.randint(2020, 2022)
    month = random.randint(1, 12)
    day = random.randint(1, 28)
    hour = random.randint(0, 23)
    minute = random.randint(0, 59)
    second = random.randint(0, 59)
    return datetime.datetime(year, month, day, hour, minute, second).strftime('%Y-%m-%d %H:%M:%S')
def generate_update_time():
    year = random.randint(2022, 2024)
    month = random.randint(1, 12)
    day = random.randint(1, 28)
    hour = random.randint(0, 23)
    minute = random.randint(0, 59)
    second = random.randint(0, 59)
    return datetime.datetime(year, month, day, hour, minute, second).strftime('%Y-%m-%d %H:%M:%S')
roles = ["后端", "前台", "运营", "产品", "前端", "测试", "财务", "运维", "销售",'主管']

data = []

for index,role in enumerate(roles,start=1):
    data.append({
        'id': index,
        'createTime':generate_create_time(),
        'updateTime': generate_update_time(),
        'roleName': role,
        'permission': []
    })
data
```

```python
with open('./roles.json','w',encoding='utf-8') as f:
    json.dump(data,f,ensure_ascii=False,indent=2)
```
