```python
import json
import time
import random
import datetime
import os
```

```python

# 随机生成用户名
s1 = ['john', 'emma', 'alex', 'sara', 'mark', 'jessica', 'awdsa', 'adam', 'olivia', 'ryan', 'chloe', 'peter', 'lucy', 'sophie', 'michael', 'emily', 'daniel', 'hannah', 'jacob', 'sophia', 'ethan']
s2 = ['123', '456', '789', 'abc', 'xyz', 'def', 'ghi', 'jkl', 'mno', 'pqr', 'stu', 'vwx', 'yz1', '234', '567']

# Generate unique usernames
usernames = []
while len(usernames) < 100:
    username = random.choice(s1) + random.choice(s2)
    if username not in usernames:
        usernames.append(username)
        
# 随机生成密码
def generate_password():
    characters = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    return ''.join(random.choices(characters, k=8))

# 随机生成电话号码
def generate_phone():
    return '1' + ''.join(random.choices('0123456789', k=10))

data = []
with open('../role/roles.json','r',encoding='utf-8') as f:
    data = json.load(f)
# 随机生成权限
def generate_role():
    roles = [i['roleName'] for i in data]
    return random.sample(roles, k=random.randint(1, 3))

# 生成随机时间
def generate_create_time():
    year = random.randint(2020, 2022)
    month = random.randint(1, 12)
    day = random.randint(1, 28)
    hour = random.randint(0, 23)
    minute = random.randint(0, 59)
    second = random.randint(0, 59)
    return datetime.datetime(year, month, day, hour, minute, second).strftime('%Y-%m-%d %H:%M:%S')
def generate_update_time():
    year = random.randint(2022, 2024)
    month = random.randint(1, 12)
    day = random.randint(1, 28)
    hour = random.randint(0, 23)
    minute = random.randint(0, 59)
    second = random.randint(0, 59)
    return datetime.datetime(year, month, day, hour, minute, second).strftime('%Y-%m-%d %H:%M:%S')


# 生成随机头像链接
def get_random_image():
    l = os.listdir('../../../static/acl/user/profilePicture')
    return f'/api/static/acl/user/profilePicture/{random.choice(l)}'
    
```

```python
# 中文姓氏列表
surnames = ['王', '李', '张', '刘', '陈', '杨', '赵', '黄', '周', '吴', '徐', '孙', '胡', '朱', '高', '林', '何', '郭', '马', '罗']

# 常见中文名字列表
name = ['明', '晓','小','芳', '军', '玲', '强', '秀英', '敏', '涛', '丽', '勇', '娜', '华', '健', '静', '刚', '燕', '杰', '红', '宇', '婷']
def f():
    return random.choice([True,False])
names1 = [i+j for i in name for j in name if i!=j]
names = random.sample(names1,len(name))+name

# 随机生成中文姓名
def generate_chinese_name():
    return random.choice(surnames) + random.choice(names)

# 其他函数保持不变

# 生成用户数据列表
def generate_user_data(num_users):
    users = []
    for i in range(num_users):
        user = {
            'createTime': generate_create_time(),
            'updateTime': generate_update_time(),
            'id': i+2,
            'name': generate_chinese_name(),  # 使用中文姓名
            'password': generate_password(),
            'phone': generate_phone(),
            'roleName': generate_role(),
            'username': usernames[i],
            'avatar': get_random_image()
        }
        users.append(user)
    return users

# 生成24条用户数据
user_data = generate_user_data(45)
user_data = [{
        'createTime': '2024-03-02 21:17:19',
        'updateTime': '2024-03-02 21:17:19',
        'id': 1,
        'name': '超级管理员',  # 使用中文姓名
        'password': '111111',
        'phone': generate_phone(),
        'roleName': ['超级管理员'],
        'username': 'admin',
        'avatar': '/api/static/acl/user/profilePicture/rabbit.jpg'
    }]+user_data
for user in user_data:
    print(user)
```
